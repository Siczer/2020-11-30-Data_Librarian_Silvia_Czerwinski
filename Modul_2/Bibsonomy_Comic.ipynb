```python
import json
```

```python
url = "https://www.bibsonomy.org/json/search/Comics?items=1000&duplicates=merged"
```

```python
import requests
```

```python
Ergebnis = requests.get(url)
```

```python
Daten = Ergebnis.json()
```

```python
items = Daten['items']
```

```python
items = [item for item in items if item['type'] == 'Publication']
```

```python
xml_file = open('bibsonomy.xml', 'w')
```

```python
xml_file.writelines("<publications>")
for Datensatz in items:
    xml_file.writelines("  <publication>")
    if "label" in Datensatz:
        xml_file.writelines ("   <title>" +Datensatz["label"]+"</title>")
    if "pub-type" in Datensatz:
        xml_file.writelines ("   <pub-type>"+Datensatz["pub-type"]+"</pub-type>" )
    if "booktitle" in Datensatz:
        xml_file.writelines ("   <booktitle>"+Datensatz["booktitle"]+"</booktitle")
    if "publisher" in Datensatz:
        xml_file.writelines ("   <publisher>"+Datensatz["publisher"]+"</publisher>")
    if "journal" in Datensatz:
        xml_file.writelines ("   <journal>"+Datensatz["journal"]+"</journal>")
    if "authors" in Datensatz:
        xml_file.writelines ("<authors>")
        for author in Datensatz["author"]:
            xml_file.writelines ("    <author>"+author+"</author>")
        xml_file.writelines (   "</authors>")
    if "year" in Datensatz:
        xml_file.writelines (   "<year>"+Datensatz ["year"]+"</year>")
    xml_file.writelines("  </publication>")
    xml_file.writelines("</publications>")
```
